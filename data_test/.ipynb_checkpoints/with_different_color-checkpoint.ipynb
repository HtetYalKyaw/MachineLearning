```python
import cv2
import numpy as np
from keras.models import load_model

# Load the pre-trained classification model
model = load_model('Waste_classify.h5')

# Define the classes
classes = ['glass', 'metal', 'paper', 'plastic']

# Define colors for each class (you can customize these)
class_colors = {
    'glass': (255, 0, 0),  # Red
    'metal': (0, 255, 0),  # Green
    'paper': (0, 0, 255),  # Blue
    'plastic': (255, 255, 0)  # Yellow
}

# Function to preprocess the frame and extract the center region
def preprocess_frame(frame):
    # Extract the center region of the frame
    height, width, _ = frame.shape
    center_x = width // 4
    center_y = height // 4
    box_width = width // 2
    box_height = height // 2
    center_frame = frame[center_y:center_y+box_height, center_x:center_x+box_width]
    
    # Resize the center region to match model input shape
    resized_frame = cv2.resize(center_frame, (224, 224))
    return resized_frame

# Function to predict class
def predict_class(frame):
    # Preprocess the frame
    preprocessed_frame = preprocess_frame(frame)
    # Resize frame to match model input shape
    resized_frame = cv2.resize(preprocessed_frame, (224, 224))
    # Expand dimensions to match model input shape
    expanded_frame = np.expand_dims(resized_frame, axis=0)
    # Make prediction
    prediction = model.predict(expanded_frame)
    predicted_class_index = np.argmax(prediction)
    predicted_class = classes[predicted_class_index]
    return predicted_class

# Function to draw bounding boxes around detected objects
def draw_boxes(frame, boxes, class_labels):
    for box, label in zip(boxes, class_labels):
        x1, y1, x2, y2 = box
        color = class_colors.get(label, (0, 0, 0))  # Default color is black if class not found in dictionary
        #cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        #Draw bounding box around the center region
        height, width, _ = frame.shape
        center_x = width // 4
        center_y = height // 4
        box_width = width // 2
        box_height = height // 2
        cv2.rectangle(frame, (center_x, center_y), (center_x + box_width, center_y + box_height), color, 2)
        cv2.putText(frame, label, (x1, y1 - 15), cv2.FONT_HERSHEY_SIMPLEX, 2, color, 4)
    return frame

# Initialize webcam
cap = cv2.VideoCapture(1)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # Preprocess the frame and extract the center region
    center_frame = preprocess_frame(frame)
    # Dummy bounding box coordinates (replace this with your object detection results)
    dummy_boxes = [ (150, 100, 450, 400)]  # Format: (x1, y1, x2, y2)
    dummy_class_labels = [predict_class(center_frame)]  # Replace this with your actual class labels
    
    # Call the function to draw bounding boxes
    frame_with_boxes = draw_boxes(frame, dummy_boxes, dummy_class_labels)
    
    # Draw dotted line box around the center region
    height, width, _ = frame.shape
    center_x = width // 4
    center_y = height // 4
    box_width = width // 2
    box_height = height // 2
    overlay = frame.copy()
    opacity = 0.1
    cv2.rectangle(overlay, (center_x, center_y), (center_x + box_width, center_y + box_height), (0, 255, 0), -1)
    cv2.addWeighted(overlay, opacity, frame, 1 - opacity, 0, frame)
    
   
    
    # Display the frame with bounding boxes
    cv2.imshow('frame', frame_with_boxes)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```
